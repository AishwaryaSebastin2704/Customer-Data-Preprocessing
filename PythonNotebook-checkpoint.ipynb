### Question 7

```python
#Reading data from processed json file
#Storing the salary and commute_distance in new variables
#Assigning the new key attribute with respective values based on if condition
#Using the sorted method in python and lambda function to sort the value in ascending order
#Writing the sorted data to new json file
with open('processed.json', 'r') as file:
    processedData = json.load(file)
for row in processedData:
    salary = row['salary']
    commute_km = row['commute_distance']
    if commute_km > 1:
        row['Salary-Commute'] = salary / commute_km
    else:
        row['Salary-Commute'] = salary
sortedData = sorted(processedData, key=lambda x: x['Salary-Commute'])
print(sortedData)
with open('sortedData.json', 'w') as output_file:
    json.dump(sortedData, output_file)
```
